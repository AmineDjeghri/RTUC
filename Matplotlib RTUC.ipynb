#### histogram 
http://www.python-simple.com/python-matplotlib/histogram.php

```python
x = [1, 2, 2, 3, 4, 4, 5, 5]
plt.hist(x, range = (0, 5), bins = 5, color = 'yellow',edgecolor = 'red')
plt.show()
```

```python
x1 = [1, 2, 2, 3, 4, 4, 4, 4, 4, 5, 5]
x2 = [1, 1, 1, 2, 2, 3, 3, 3, 3, 4, 5, 5, 5]
bins = [x + 0.5 for x in range(0, 6)]
plt.hist([x1, x2], bins = bins, color = ['yellow', 'green'],
            edgecolor = 'red', hatch = '/', histtype = 'bar',orientation = 'horizontal') 
plt.show()
```

#### bars
http://www.python-simple.com/python-matplotlib/barplot.php

```python
barWidth = 0.4
y1 = [1, 2, 4, 3]
y2 = [3, 4, 4, 3]
r1 = range(len(y1))
r2 = [x + barWidth for x in r1]

plt.bar(r1, y1, width = barWidth, color = ['red' for i in y1],
           edgecolor = ['blue' for i in y1], linewidth = 2)
plt.bar(r2, y2, width = barWidth, color = ['pink' for i in y1],
           edgecolor = ['black' for i in y1], linewidth = 4)
plt.xticks([r + barWidth / 2 for r in range(len(y1))], ['A', 'B', 'C', 'D'])
plt.show()
```

```python
#plt.subplot(n_rows,n_columns,i)  n_rows * n_columns must be equal to numer of plots (here it's 3), index= number of subplot

plt.subplot(1,3,1)
plt.plot(data)

plt.subplot(1,3,2) 
plt.plot(data)
plt.xlim([0,5])
plt.title('plot 2')

plt.subplot(1,3,3)
plt.plot(data)

plt.show()
```

```python
for i in range(6):
    #plt.subplot(n_rows,n_columns,i)  n_rows * n_columns must be equal to numer of plots (here it's 6)
    plt.subplot(2,3,i+1)
    plt.plot(data)
plt.show() #this must be called only once when subplotting
```

```python
#subplotting with differents datas, lables and titles
datas=[data,data1,data2]
titles=['title1','title2','title3']
x_labels=['x1','x2','x3']
y_labels=['y1','y2','y3']

for i in range(3):
    #plt.subplot(n_rows,n_columns,i)   n_rows * n_columns must be equal to numer of plots (here it's 6)
    plt.subplot(1,3,i+1)
    plt.plot(datas[i])
    
    plt.xlabel(x_labels[i])
    plt.ylabel(y_labels[i])
    plt.title(titles[i])
    
plt.tight_layout()
plt.show() #this must be called only once when subplotting
```

#### 2- background styling: , use `with plt.style.context`

```python
print(plt.style.available)
print(len(plt.style.available))
```

```python
#plot with temporary styling
with plt.style.context('seaborn-whitegrid'):
    plt.plot(data1)
plt.show() 

#plot with default
plt.plot(data1)
plt.show()

#plot with temporary styling
with plt.style.context('dark_background'):
    plt.plot(data1)
plt.show() 
```

### 2-permanent styling
 If you want to change the default global styling 

#### to change the default background: use `plt.style.use`

```python
plt.plot(data1)
plt.show()

plt.style.use('seaborn-whitegrid')
plt.plot(data1)
plt.show()
```

#### 2- if you want to change the default settings of: lines, color ..ect 
#### By default, each line is assigned a different color and a style specified by a 'color / style cycle’. 
#### To change this behavior, you can edit the `axes.prop_cycle`  and the `axes.prop_cycle` rcParam.

```python
#lines 
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['lines.linestyle'] = '--'
plt.plot(data1)
```

```python
# colors
plt.plot(data1) #blue is the first default color
plt.plot(data2) #orange is the second default color
plt.show()

from cycler import cycler
plt.rcParams['axes.prop_cycle'] = cycler(color=['r', 'g', 'b', 'y'])

plt.plot(data1) #r for blue will be the first default color
plt.plot(data2) #g for green will be the second default color
plt.show()
```

#### 3- if you want to reset everything to the default settings

```python
plt.rcParams.update(plt.rcParamsDefault)
%matplotlib inline
```

```python
plt.style.use('seaborn-whitegrid')
plt.plot(data1)
plt.show()

plt.rcParams.update(plt.rcParamsDefault)
%matplotlib inline

plt.plot(data1)
plt.show()
```
