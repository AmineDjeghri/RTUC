## 2- Styling

```python
data1= np.random.randn(10)
data2= np.random.randn(10)
```

### 1-Temporary styling : 
If you only want to use a style for a specific block of code but don't want to change the global default styling


#### 1- lines, colors, markers ...ect

```python
plt.plot(data1, color='green', linestyle='dashed', marker='o',markerfacecolor='blue', markersize=12)
plt.show()
plt.plot(data1)
plt.show()
```

#### 2- background styling: , use `with plt.style.context`

```python
print(plt.style.available)
print(len(plt.style.available))
```

```python
#plot with temporary styling
with plt.style.context('seaborn-whitegrid'):
    plt.plot(data1)
plt.show() 

#plot with default
plt.plot(data1)
plt.show()

#plot with temporary styling
with plt.style.context('dark_background'):
    plt.plot(data1)
plt.show() 
```

### 2-permanent styling
 If you want to change the default global styling 

#### to change the default background: use `plt.style.use`

```python
plt.plot(data1)
plt.show()

plt.style.use('seaborn-whitegrid')
plt.plot(data1)
plt.show()
```

#### 2- if you want to change the default settings of: lines, color ..ect 
#### By default, each line is assigned a different color and a style specified by a 'color / style cycle’. 
#### To change this behavior, you can edit the `axes.prop_cycle`  and the `axes.prop_cycle` rcParam.

```python
#lines 
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['lines.linestyle'] = '--'
plt.plot(data1)
```

```python
# colors
plt.plot(data1) #blue is the first default color
plt.plot(data2) #orange is the second default color
plt.show()

from cycler import cycler
plt.rcParams['axes.prop_cycle'] = cycler(color=['r', 'g', 'b', 'y'])

plt.plot(data1) #r for blue will be the first default color
plt.plot(data2) #g for green will be the second default color
plt.show()
```

#### 3- if you want to reset everything to the default settings

```python
plt.rcParams.update(plt.rcParamsDefault)
%matplotlib inline
```

```python
plt.style.use('seaborn-whitegrid')
plt.plot(data1)
plt.show()

plt.plot(data1)
plt.show()
```
